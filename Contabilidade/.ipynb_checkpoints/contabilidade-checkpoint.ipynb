```python
estado = geobr.read_state()
estado.head(28)
```

```python

# Specify the URL of the CSV file
url = "C:/users/gusta/contabilidade/resultados.csv"

# Read the CSV file specifying the delimiter, encoding, and quoting
d = pd.read_csv(url, on_bad_lines='skip')

# Display the first 27 rows
d.head(27)
```

```python
ano_desejado = 2019  # Aqui você define o ano desejado

# Filtrar os dados para o ano desejado
d_ano_desejado = d[d['Ano'] == ano_desejado]

# Selecionar apenas as colunas desejadas
d_ano_desejado = d_ano_desejado[["UF", "Média Empenhado"]]
d_ano_desejado = d_ano_desejado.rename(columns={"UF":"abbrev_state"})
```

```python
estado_d = estado.merge(d_ano_desejado)
estado_d = estado_d[["geometry", "Média Empenhado"]]
```

```python
f, ax = plt.subplots()
f.set_size_inches(16, 16)
estado_d.plot(
    ax=ax, # Axis de destino do gráfico
    column="Média Empenhado", # Coluna com os valores usados para colorir o mapa
    cmap="Reds", # Mapa de cores
    edgecolor="black", # Cor dos contornos
    linewidth=0.25, # Espessura dos contornos
)

cax = f.add_axes(
    [
        0.82,    # posicao x (entre 0.0 e 1.0)
        0.18,    # posicao y (entre 0.0 e 1.0)
        0.03,    # largura x
        0.40,    # altura y
    ]
)

sm = plt.cm.ScalarMappable(
    cmap="Reds",                                       # Usa o mesmo cmap do mapa
    norm=plt.Normalize(
        vmin=estado_d["Média Empenhado"].min(),  # Valor mínimo
        vmax=estado_d["Média Empenhado"].max(),  # Valor máximo
    ),
)
# Põe o Axis com Colorbar na mesma figura do mapa
f.colorbar(
    sm,
    cax=cax,
    # Formata a escala do Colorbar
    format=ticker.FuncFormatter(lambda x, pos: f"{x/1: >10,.0f}  R$"),
)

# Adiciona um título ao mapa
ax.set_title(
    "Análise de ações financeiras da saúde pública brasileira",
    fontdict={"fontsize": 20},
)

# Adiciona a fonte como nota de rodapé
f.text(
    0.15,                  # Posição x
    0.20,                  # Posição y
    "Fonte: IBGE (2022)."  # Texto
)

ax.axis("off")          # Remove os eixos
```
